## Data Cleaning

```python
# Checking for missing values

print("Missing Values in leads")
print(leads.isnull().sum())

print("\n Missing values in deals")
print(deals.isnull().sum())
```

```python
#Dropping Columns with missing values in deals
deals= deals.drop(columns=['has_company','has_gtin','average_stock','declared_product_catalog_size','lead_behaviour_profile'])
print("Deals after dropping:",deals.shape)
```

```python
# Drop rows with missing values in leads origin
leads = leads.dropna(subset=['origin'])

# Drop rows with missing values in deals
deals = deals.dropna(subset=['lead_type', 'business_segment'])

print("Leads shape after cleaning:", leads.shape)
print("Deals shape after cleaning:", deals.shape)
```

```python
# Fill missing business_type with unknown
deals['business_type'] = deals['business_type'].fillna('unknown')

print("Missing values in deals:", deals.isnull().sum())
```

```python
print("Missing Values in leads")
print(leads.isnull().sum())
```

```python
# Check for duplicates
print("Duplicate leads:", leads.duplicated().sum())
print("Duplicate deals:", deals.duplicated().sum())
```

```python
# Check data types
print(leads.dtypes)
print("\n")
print(deals.dtypes)
```

```python
# Fixing date columns
leads['first_contact_date'] = pd.to_datetime(leads['first_contact_date'])
deals['won_date'] = pd.to_datetime(deals['won_date'])

print(leads['first_contact_date'].dtype)
print(deals['won_date'].dtype)
```

```python
# Checking for outliers
print(deals['declared_monthly_revenue'].describe())
```

```python
Q1 = deals['declared_monthly_revenue'].quantile(0.25)
Q3 = deals['declared_monthly_revenue'].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
outliers = deals[deals['declared_monthly_revenue'] > upper]
print("Revenue outliers found:", len(outliers))
print("Note: Outliers kept as they represent genuine high-revenue businesses")
```

```python
#Checking for inconsistent values
print(leads['origin'].unique())
print(deals['business_segment'].unique())
```

## Merging Datasets

```python
df = pd.merge(leads, deals, on='mql_id', how='left')
print("Merged dataset shape:", df.shape)
```

## Data Analysis

### Funnel Overview

```python
# Overall Funnel Analysis
total_leads = len(leads)
total_converted = len(deals)
conversion_rate = round((total_converted / total_leads) * 100, 2)
not_converted = total_leads - total_converted

print("=== MARKETING FUNNEL OVERVIEW ===")
print(f"Total Leads: {total_leads}")
print(f"Converted to Deals: {total_converted}")
print(f"Not Converted: {not_converted}")
print(f"Conversion Rate: {conversion_rate}%")
print(f"Drop-off Rate: {round(100 - conversion_rate, 2)}%")
```

### Channel Performance Analysis

```python
channel_leads = leads['origin'].value_counts()
channel_deals = df[df['won_date'].notna()]['origin'].value_counts()
channel_conversion = round((channel_deals / channel_leads * 100), 2)

print("=== CHANNEL PERFORMANCE ===")
print("\nLeads per channel:")
print(channel_leads)
print("\nDeals per channel:")
print(channel_deals)
print("\nConversion rate per channel:")
print(channel_conversion.sort_values(ascending=False))
```

### Business Segment Performance

```python
segment_performance = df[df['won_date'].notna()]['business_segment'].value_counts()
print("=== TOP BUSINESS SEGMENTS ===")
print(segment_performance.head(10))
```

### Monthly Trend Analysis

```python
df['won_month'] = pd.to_datetime(df['won_date']).dt.to_period('M')
monthly_conversions = df[df['won_date'].notna()]['won_month'].value_counts().sort_index()

print("=== MONTHLY CONVERSIONS ===")
print(monthly_conversions)
```

### Lead Type Analysis

```python
lead_type_performance = df[df['won_date'].notna()]['lead_type'].value_counts()
print("=== LEAD TYPE PERFORMANCE ===")
print(lead_type_performance)
```

## Visualization

### Funnel Chart

```python
plt.figure(figsize=(10,6))
funnel_stages = ['Total Leads', 'Converted', 'Not Converted']
funnel_values = [total_leads, total_converted, not_converted]
plt.bar(funnel_stages, funnel_values, color=['steelblue','lightgreen','red'])
plt.title('Marketing Funnel Overview')
plt.ylabel('Number of Leads')
plt.tight_layout()
plt.show()
```

### Channel Conversion Rates 

```python
plt.figure(figsize=(12,6))
channel_conversion.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Conversion Rate by Marketing Channel')
plt.xlabel('Channel')
plt.ylabel('Conversion Rate %')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Top Business Segments

```python
plt.figure(figsize=(12,6))
segment_performance.head(10).plot(kind='barh', color='teal')
plt.title('Top 10 Business Segments by Deals Closed')
plt.xlabel('Business Segment')
plt.ylabel('Number of Deals')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Monthly Trends

```python
plt.figure(figsize=(15,6))
plt.plot(monthly_conversions.index.astype(str), 
         monthly_conversions.values, 
         color='purple', linewidth=2, marker='o')
plt.title('Monthly Conversion Trends')
plt.xlabel('Month')
plt.ylabel('Number of Conversions')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
